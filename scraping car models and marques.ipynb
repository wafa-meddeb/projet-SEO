```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time

# Set up Chrome options
chrome_options = Options()
chrome_options.add_argument("--headless")  # To run Chrome in headless mode

# Set up the Chrome driver
chrome_driver_path = 'C:/Users/ASUS/Downloads/chromedriver-win64/chromedriver-win64/chromedriver.exe'  # Update with your Chrome driver path
service = Service(chrome_driver_path)
driver = webdriver.Chrome(service=service, options=chrome_options)

# Open the website
url = "https://www.caradisiac.com/constructeurs--automobiles/"
driver.get(url)

# Wait for the page to load
time.sleep(5)  # Adjust the sleep time as needed

# Define the alphabet range from 'A' to 'Z'
alphabet_range = [chr(i) for i in range(ord('A'), ord('Z') + 1)]

# Initialize an empty list to store car makes
car_makes = []

# Loop through each letter in the alphabet range
for letter in alphabet_range:
    # Construct the ID dynamically
    list_id = f"list_{letter}"
    
    # Check if the element with the constructed ID exists
    try:
        car_make_div = driver.find_element(By.ID, list_id)
        
        # Find all the <a> elements within the div
        car_make_links = car_make_div.find_elements(By.TAG_NAME, 'a')
        
        # Extract the marque names from the 'title' attribute of each <a> element
        car_makes.extend([link.get_attribute('title') for link in car_make_links])
    except:
        pass

# Print the list of car makes
for make in car_makes:
    print(make)

# Close the browser
driver.quit()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def scrape_models(brand_name):
    models = []
    url_template = "https://www.caradisiac.com/auto--{}/modeles/"
    brand_link = url_template.format(brand_name)

    response = requests.get(brand_link)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        a_elements = soup.select('.txtModeleGalerie a')
        for a in a_elements:
            model_name = a.get('title', '').strip()
            if model_name:
                models.append(model_name)
    else:
        print(f"Failed to fetch for {brand_name}: {response.status_code}")
    return models

# Example brand list - replace 'df["brands"]' with your actual list of brands
brand_names = brand_names

# List to hold all data
data = []

# Iterate over each brand and scrape models
for brand_name in brand_names:
    models = scrape_models(brand_name)
    for model in models:
        # Extract model name without the brand
        model_without_brand = model.replace(brand_name.capitalize(), '').strip()
        data.append({"brand": brand_name.capitalize(), "model": model, "model_without_brand": model_without_brand})
#The capitalize() function in Python converts the first character of a string to uppercase and the remaining characters to lowercase.
# Create DataFrame
df = pd.DataFrame(data, columns=["brand", "model", "model_without_brand"])

# Display the DataFrame
print(df)
```
