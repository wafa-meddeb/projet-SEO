```python
import pandas as pd
#from fbprophet import Prophet
#import matplotlib.pyplot as plt

# Load the data
file_path = "C:/Users/ASUS/Desktop/stage_pfe/pred_data.csv"
data = pd.read_csv(file_path)

# Drop the 'Unnamed: 0' column
data = data.drop(columns=['Unnamed: 0'])

# Melt the dataframe to have a single 'ds' column for dates and a 'y' column for values
data_melted = data.melt(var_name='Date', value_name='Searches', ignore_index=False)
data_melted.reset_index(inplace=True)

# Remove the 'Searches: ' prefix and the extra year part
data_melted['Date'] = data_melted['Date'].str.replace('Searches: ', '').str.strip()

# Convert the 'Date' column to datetime format
data_melted['Date'] = pd.to_datetime(data_melted['Date'], format='%b %Y')

# Rename columns to fit Prophet requirements
data_melted.rename(columns={'Date': 'Date', 'Searches': 'Searches'}, inplace=True)


```

```python
data_melted
```

```python
data_melted.drop(columns="index")
```

```python
# Set the 'Date' column as the index
data_melted.set_index('Date', inplace=True)
```

```python
data_melted
```

```python
data_melted.drop(columns="index")
```

```python
# Apply log transformation
import numpy as np
data_melted['Searches'] = data_melted['Searches'].apply(lambda x: np.log1p(x))
```

```python
data_melted
```

```python
data_melted["Searches"].isna().sum()
```

```python
# Visualize the data before handling NaNs
plt.figure(figsize=(10, 5))
plt.plot(data_melted.index, data_melted['Searches'], marker='o', linestyle='-', color='blue', label='Original Data')
plt.title('Searches Data with NaNs')
plt.xlabel('Date')
plt.ylabel('Searches')
plt.legend()
plt.show()
```

```python
# Interpolate missing values
data_melted.interpolate(method='spline', order= 3,inplace=True)
# Ensure no NaN values are present
data_melted.fillna(method='ffill', inplace=True)  # Forward fill any remaining NaNs
data_melted.fillna(method='bfill', inplace=True)  # Backward fill if needed
```

```python
# Visualize the data after handling NaNs
plt.figure(figsize=(10, 5))
plt.plot(data_melted.index, data_melted['Searches'], marker='o', linestyle='-', color='orange', label='spline Interpolated Data')
plt.title('Searches Data after Interpolation')
plt.xlabel('Date')
plt.ylabel('Searches')
plt.legend()
plt.show()
```
