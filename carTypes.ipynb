```python
data = pd.DataFrame()
data["models"] = models_list
```

```python
data['Tokenized_models'] = data['models'].apply(lambda x: text_to_word_sequence(x))
```

```python
# Function to check if any car type exists in the model
def check_car_type(tokenized_models):
    for word in tokenized_models:
        if word in car_types_list:
            return word
    return "does not exist"

# Apply the function to create the 'car_types' column
data['car_types'] = data['Tokenized_models'].apply(check_car_type)

```

```python
# Function to assign car types to models
def assign_car_type(model):
    for car_type, models in car_type_dict.items():
        if model in models:
            return car_type
    return "unknown"

# Apply the function to create the 'car_types' column
types_not_found_df['car_types'] = types_not_found_df['models'].apply(assign_car_type)
```
