```python
# Define a function to set the 'category' based on the value of the last three columns
def set_category(data):
    if data['générique_générique'] == 1:
        return 'générique'
    elif data['Marque_marque'] == 1:
        return 'marque'
    elif data['modèle_modèle'] == 1:
        return 'modèle'
    else:
        return 'unknown'

# Apply the function to each row to set the 'category' column
df['category'] = df.apply(set_category, axis=1)
```

**Models**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier


# Defining the models to train
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Random Forest": RandomForestClassifier(),
    "SVM": SVC(),
    "KNN": KNeighborsClassifier()
}

```

```python
y_train
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
# Training and evaluating the models
for name, model in models.items():
    # Create a pipeline that first vectorizes the text and then applies the model
    pipeline = make_pipeline(TfidfVectorizer(), model)
    
    # Train the model
    pipeline.fit(X_train, y_train)

    # Make predictions on the test set
    y_pred = pipeline.predict(X_test)
    
    # Calculate the accuracy
    accuracy = accuracy_score(y_test, y_pred)
    
    # Print the model name and its accuracy
    print(f"Model: {name}")
    print(f"Accuracy: {accuracy:.4f}")
    
    # Print confusion matrix
    print(f'Confusion matrix of {name}:\n', confusion_matrix(y_test, y_pred))
    print()
```
