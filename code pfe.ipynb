**PREPARING THE CORPUS FOR MODEL FINE-TUNING**

**Keywords Tokenisation**

```python
from keras.preprocessing.text import text_to_word_sequence
dataset = pd.DataFrame()
dataset['Keyword'] = df['Keyword']
# Apply text_to_word_sequence function to each keyword and store the tokens in the Tokenized_Keyword column
dataset['Tokenized_Keyword'] = df['Keyword'].apply(lambda x: text_to_word_sequence(x))
```

**POS tags**

```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

from nltk.tokenize import word_tokenize

# Function to add POS tags to tokens
def add_pos_tags(tokens):
    pos_tags = nltk.pos_tag(tokens)
    return pos_tags
```

```python
corpus.loc[:,'POS_Tags'] = corpus.loc[:, 'Tokenized_Keyword'].apply(add_pos_tags)
```

```python
corpus
```

**Deleting stopwords**

```python
from nltk.corpus import stopwords
fr_stops = set(stopwords.words('french'))

def remove_stopwords(tokens):
    return [token for token in tokens if token.lower() not in fr_stops]

# Remove stopwords from tokenized keywords
corpus.loc[:, 'Tokenized_Keyword'] = corpus.loc[:, 'Tokenized_Keyword'].apply(remove_stopwords)
```

**word2Vec**

```python
#!pip install gensim
from gensim.utils import simple_preprocess
from gensim.models import Word2Vec
```

```python
#  CBOW model training
model_cbow = Word2Vec(sentences = corpus["Tokenized_Keyword"], vector_size=100, window=5, min_count=2, sg=0)

# Skip Gram model training
model_skipgram = Word2Vec(sentences = corpus["Tokenized_Keyword"], vector_size=100, window=5, min_count=2, sg=1)
```

```python
# choose a word
target_word = 'mercedes'

# most similar words with CBOW
similar_words_cbow = model_cbow.wv.most_similar(target_word)
print(f'CBOW Similar Words to "{target_word}": {similar_words_cbow}')

# most similar words with Skip Gram
similar_words_skipgram = model_skipgram.wv.most_similar(target_word)
print(f'Skip Gram Similar Words to "{target_word}": {similar_words_skipgram}')
```
